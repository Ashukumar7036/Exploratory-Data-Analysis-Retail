## > Most of the categories belong to Office Supplies 
## > 20-25% belongs to furniture and technology.

```python
store.pivot_table(values='Sales', index='Segment', columns='Discount', aggfunc='median')
```

```python
store.pivot_table(values='Profit',index='Segment',columns='Discount',aggfunc='median')
```

### From the above table -
For Discount more than or equal to 30% the Superstore was facing losses.
For Discount b/w 0-20% the Superstore was making profits.
